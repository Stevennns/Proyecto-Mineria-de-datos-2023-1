### Características del Dataset

El dataset obtenido del sitio web contiene tres carpetas: **annotations**, **json** y **data_image**. 
La carpeta *annotations* corresponde a archivos *.tsv* con información medida, verificada y etiquetada manualmente acerca de cada tweet.<br/>
En *data_image* se encuentran las rutas a las fotos extraídas de los tweet. ### Propuesta experimental

#### **¿Existen características en la forma en la que se generan tweets durante una crisis que estén relacionadas con la magnitud del desastre?**

Aqui se trato de responder a la primera pregunta. Como primer paso, es necesario cuantificar la magnitud del desastre. Para lograr esto, calcularemos un indicador que tomará en cuenta la categoría de los tweets, la cual está contenida en el atributo "text human".<br/>
El indicador se calcula de la siguiente manera: primero, se filtran los tweets para mantener únicamente aquellos que sean de categoría informativa y relevante. Luego, se cuenta el número de tweets para cada categoría humanitaria y se normaliza en función del número total de tweets. Esto nos proporciona una medida del impacto del desastre en cada aspecto humanitario. A continuación, se ponderan estas categorías utilizando sus respectivos pesos:

* "Affected individuals": 0.2
* "Infrastructure and utility damage": 0.2
* "Injured or dead people": 0.3
* "Missing or found people": 0.05
* "Not humanitarian": 0
* "Rescue, volunteering or donation effort": 0.05
* "Vehicle damage": 0.05
* "Other relevant information": 0.15

Estos valores ponderados se suman, lo que resulta en el indicador de magnitud del desastre. Este valor puede interpretarse como una estimación de la proporción de tweets informativos y relevantes que reflejan el impacto del desastre en diferentes aspectos humanitarios.

Esta magnitud se agregó al dataframe en una columna llamada "magnitud" igual a todos los tweets de un mismo desastre.<br/>
Cabe destacar que se eliminaron los registros del desastre de Sri Lanka y aquellos registros que eran NaN.<br/>
Luego de agregar la columna _magnitud_ al dataframe, se aplicó clustering sobre los registros. Para ello, primeramente, se conservaron las siguientes columnas:

   * favorite_count  
   * retweet_count  
   * reply_count  
   * quote_count  
   * tdd  
   * magnitud

Tras esto, se preprocesaron los registros usando la librería scikit-learn y se utilizó el método del codo para encontrar los números de clusters óptimos usando **K-Means** como método de clustering. El resultado fue 3 clusters como número óptimo. Luego, se aplicó el algoritmo _K-Means_ nuevamente con 3 clusters.
